```python
# Q1 Basic Data Exploration
import pandas as pd

# Load the dataset
df = pd.read_csv('Salaries.csv')

# Display number of rows and columns
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

# Display data types of each column
print("\nData types of each column:")
print(df.dtypes)

# Check for missing values in each column
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Q2 Descriptive Statistics
import pandas as pd

# Load the dataset
df = pd.read_csv('Salaries.csv')

# Calculate basic statistics
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
salary_range = max_salary - min_salary
std_dev_salary = df['TotalPay'].std()

# Print results
print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
print("Minimum Salary:", min_salary)
print("Maximum Salary:", max_salary)
print("Salary Range:", salary_range)
print("Standard Deviation of Salary:", std_dev_salary)
```

```python
# Q3 Data Cleaning
# Impute missing values with mean in 'BasePay' and 'Benefits' columns
mean_basepay = df['BasePay'].mean()
df['BasePay'].fillna(mean_basepay, inplace=True)
mean_Benefits = df['Benefits'].mean()
df['Benefits'].fillna(mean_Benefits, inplace=True)

# Drop missing values in 'Notes' and 'Status' columns
df.dropna(subset=['Notes', 'Status'], inplace=True)
```

```python
# Q4 Basic Data Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Salaries.csv')

# Visualize the distribution of salaries with a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Create a pie chart to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(10)  # Select top 10 departments
plt.figure(figsize=(10, 6))
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Top 10 Departments')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Q5 Grouped Analysis
import pandas as pd

# Load the dataset
df = pd.read_csv('Salaries.csv')

# Group the data by one or more columns and calculate summary statistics
grouped_data = df.groupby('JobTitle')  # Group by job title
# Calculate average salary for each group
average_salary_by_job = grouped_data['TotalPay'].mean()

# Print summary statistics
print("Average Salary by Job Title:")
print(average_salary_by_job)

# Compare the average salaries across different groups visually (if desired)
# For example, you can create a bar plot
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
average_salary_by_job.sort_values().plot(kind='barh', color='skyblue')
plt.title('Average Salary by Job Title')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.show()
```

```python
# Q6 Simple Correlation Analysis
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Salaries.csv')

# Identify correlation between salary and another numerical column (e.g., BasePay)
correlation = df['TotalPay'].corr(df['BasePay'])

# Print the correlation coefficient
print("Correlation between TotalPay and BasePay:", correlation)

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df['BasePay'], df['TotalPay'], alpha=0.5, color='skyblue')
plt.title('Scatter Plot of TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.grid(True)
plt.show()
```
